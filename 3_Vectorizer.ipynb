```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
import gensim
import numpy as np
import pandas as pd
from joblib import dump, load
```

```python
df_train = pd.read_csv("X_train_stemm_cleaned.csv")
df_valid = pd.read_csv("X_valid_stemm_cleaned.csv")
df_test = pd.read_csv("X_test_stemm_cleaned.csv")

X_train_clean = df_train['tweet_stemming']
X_valid_clean = df_valid['tweet_stemming']
X_test_clean = df_test['tweet_stemming']

y_train = pd.read_csv("y_train.csv")
y_valid = pd.read_csv("y_valid.csv")
y_test = pd.read_csv("y_test.csv")
```

```python
# Load model TF-IDF
tfidf_vectorizer = load('tfidf_vectorizer.joblib')

# transform set data dari load model tfidf yang telah disimpan
X_train_tfidf = tfidf_vectorizer.transform(X_train_clean)
X_valid_tfidf = tfidf_vectorizer.transform(X_valid_clean)
X_test_tfidf = tfidf_vectorizer.transform(X_test_clean)
```

# word2vec Skipgram

```python
# Tokenisasi data untuk Word2Vec
all_sentences = [word_tokenize(text) for text in X_train_clean]

# Training model Word2Vec dengan data train
W2V_SIZE = 75
W2V_WINDOW = 7
W2V_EPOCH = 32
W2V_MIN_COUNT = 7
word2vec_model = gensim.models.word2vec.Word2Vec(vector_size=W2V_SIZE,
                                            window=W2V_WINDOW,
                                            min_count=W2V_MIN_COUNT,
                                            sg = 1,
                                            workers=8)
word2vec_model.build_vocab(all_sentences)
word2vec_model.train(all_sentences, total_examples=word2vec_model.corpus_count, epochs=W2V_EPOCH)
```

```python
# save model
word2vec_model.save("w2v_skipgram.model")
```
